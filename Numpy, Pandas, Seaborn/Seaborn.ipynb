# Uzduotys


```python
mpg = sns.load_dataset('mpg')
mpg.head()
```

```python
mpg["l_100km"] = round(235.21 / mpg["mpg"],1)
mpg.head(10)
```

```python
mpg["kubatura l."] = round(mpg["displacement"] * 0.0163871, 1)
```

```python
mpg.head(10)
```

```python
# 1 Atspausdinkite histogramą, kurioje matytųsi, 
# kiek automobilių turi kokią akseleraciją. 
# Šioje ir kitose užduotyse žaiskite su stiliais ir spalvomis, taip, kaip jums patinka.

sns.displot(mpg['acceleration'], bins=20)
```

```python
# 2 Atspausdinkite histogramą, kurioje matytųsi, 
# kiek automobilių turi kokius variklio tūrius.

sns.displot(mpg["kubatura"])
```

```python
# 3 Atspausdinkite histogramą, kurioje matytųsi, 
# kokie yra cilindrų skaičiaus variantai.

sns.countplot(x='cylinders', data=mpg)
```

```python
# 4 Atspausdinkite histogramą, kurioje matytųsi, 
# kiek yra pagaminimo metų variantų

sns.countplot(x='model_year', data=mpg)
```

```python
# 5 Atspausdinkite histogramą, kurioje matytųsi, 
# kiek automobilių lentelėje kokia šalis pagamino.

sns.countplot(x='origin', data=mpg)
```

```python
# 6 Atspausdinkite histogramą, kurioje matytųsi, 
# koks kurioje šalyje pagamintų automobilių variklio tūrio vidurkis.

sns.barplot(x="origin", y="displacement", data=mpg, ci=False)
```

```python
# 7 Atspausdinkite sklaidos diagramą, 
# kurios x ašis būtų 'displacement', y - 'acceleration', 
# taip pat kiekvienas taškas atspindėtų šalį gamintoją ir cilindrų skaičių

sns.scatterplot(x='displacement', y='acceleration', data=mpg, hue="origin", size="cylinders", palette="Set1")

# sns.barplot(x='sex', y='total_bill', data=tips, hue='day', estimator=sum)
```

```python
# 8 Atspausdinkite visas įmanomas sklaidos diagramas lentelėje, 
# kur pagal taško spalvą matytumėm šalį gamintoją. 
# Kokias tendencijas galima aiškiai išskirti?

sns.pairplot(data=mpg, hue="origin") # diag_kind="hist")
```

```python
# 9 Atspausdinkite stulpelinę diagramą, 'origin' x 'mpg'. 
# Pabandykite interpretuoti rezultatą.

sns.boxplot(x="origin", y="mpg", data=mpg)
```

```python
# 10 Sukurkite koreliacijų matricą. 
# Jos pagrindu atspausdinkite mozaikinę diagramą.

koreliacija = mpg.corr(numeric_only=True)
```

```python
koreliacija
```

```python
sns.heatmap(koreliacija, annot=True, cmap="cividis") 
```

```python
# 11 Atspausdinkite sklaidos diagramų rinkinį, 
# kuriame kiekviena lentelė pagal šalį rodytų 'acceleration' ir 'mpg' sąntykį.

data = sns.FacetGrid(data=mpg, col="origin")
data.map(sns.scatterplot,"acceleration","cylinders")

```

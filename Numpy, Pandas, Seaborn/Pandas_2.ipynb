# Trūkstami duomenys

```python
# pirmas metodas tvarkytis su trūkstamais duomenimis yra .dropna(). 
# Jis išmeta visas eilutes, kuriose yra bent viena NaN reikšmė:


table.dropna()
```

```python
# jeigu norime išmesti stulpelius, turinčius bent vieną NaN, turime nurodyti ašį:
table.dropna(axis=1)
```

```python
# Galime išmesti tik tas eilutes, ar stulpelius, kurie turi mažiau, negu (pvz) 5 NE NaN reikšmes:
table.dropna(thresh=5)

# Papildomai nurodžius ašį, tą patį galima padaryti ir su stulpeliais.
```

```python
# Jei norime NaN reikšmes pakeisti kažkuo kitu, naudosime .fillna() metodą:

table.fillna('bupkis')
```

```python
# Daugiau prasmės būtų pakeisti pvz. stulpelio vidurkiu:

table['W'].fillna(value=table['W'].mean())
```

```python
table.fillna(value=table.mean())
```

```python
# .describe() metodas duoda pagrindinę informaciją apie lentelės duomenis:

baltic.describe()
```

```python
# Jeigu data_main ir data_right turėtų po vieną sutampantį stulpelį, pvz.:

key = 'A B C D E F'.split()
data_main['key'] = key
data_right['key'] = key
```

```python
data_right
```

```python
# galima jas sulieti to bendro stulpelio pagrindu:

pd.merge(data_main, data_right, on='key')

# matome, kad lentelės suklijuotos, tačiau stulpelis 'key' nesikartoja (stulpeliai pervadinti, kad nesidubliuotų pavadinimai). 
# Stulpelį 'key' teko nurodyti parametre on.
```

```python
data_main
```

```python
pd.merge(data_main, data_right, on=['key', 'key2'])

# Matome, kad į parametrą on galima perduoti ir daugiau reikšmių, sąraše.
```

```python
# Yra ir metodas .join(). Jis daro tą patį, ką ir merge, tik indekso pagrindu. Pvz.:

data_main.join(data_right, lsuffix='.L', rsuffix='.R')

# Kadangi sutampa stulpelių pavadinimai, teko nurodyti lsuffix ir rsuffix. 
# Jeigu indeksas nebūtų identiškas, rezultate gautumėm tik tas eilutes, kurių indeksas persidengia. 
# Dažniausiai vis tik naudosime .merge()
```

# Uzduotys

```python
# 1 Sulipdykite iš 4 fragmentų vieną lentelę:

top_left = pd.read_csv('top1-25-1.csv')
top_right = pd.read_csv('top1-25-2.csv')
bottom_left = pd.read_csv('top26-50-1.csv')
bottom_right = pd.read_csv('top26-50-2.csv')

# Pirmiausiai sujunkite apatinę dalį, tada viršutinę (yra persidengiančių stulpelių!), galiausiai viršų su apačia.
# Naudokite .concat() ir .merge() metodus.
```

```python
bottom = pd.concat([bottom_left, bottom_right], axis=1)
```

```python
top = pd.merge(top_left, top_right, on=["Track.Name", "Popularity"])
```

```python
data = pd.concat([top, bottom], sort=False)
data.head()
```

```python
# 2 Sutvarkykite indeksą - padarykite, kad prasidėtų nuo 1.

data["#"] = list(range(1, 51))
data.set_index("#", inplace=True)
data.head()
```

```python
# 3 Sukurkite grupavimo pagal žanrą objektą

data.groupby("Genre")
genre = data.groupby("Genre")
```

```python
# 4 kokie žanrai lentelėje pasitaiko daugiau negu 3 kartus?

df = genre.count()
df[df["Energy"] > 3]["Energy"]
```

```python
# 5 Koks žanras pats populiariausias? Koks mažiausiai populiarus?
most_pop_genre = genre["Popularity"].mean().idxmax()
most_pop_score = genre["Popularity"].mean().max()
least_pop_genre = genre["Popularity"].mean().idxmin()
least_pop_score = genre["Popularity"].mean().min()
print(most_pop_genre, most_pop_score)
print(least_pop_genre, least_pop_score)

```

```python
# 6 Sukurkite lentelę, kurioje matytųsi, koks žanras turi aukščiausią vidurkį kiekviename indikatoriuje, 
# bei pats vidurkis.
# Kad nereikėtų daug vargti su stulpelių pavadinimais, 
# galima juos ištraukti su .columns.tolist()

column = data.columns.tolist()[3:]
zanrai = genre[column].mean().idxmax()
sk = genre[column].mean().max()
ats = pd.DataFrame([column, zanrai, sk], ["Indicator", "Genre", "Points"]).transpose()
ats
```

```python
# 7 Grįžkime prie sulipdytos lentelės. 
# Ištraukite visas eilutes, kurios turi NaN reikšmių.
# filtravimo sąlygose naudokite .isnull()* funkciją.*

data.info()
data[data["Genre"].isnull() | data["Popularity"].isnull()]
```

```python
# 8 Stulpelyje 'Genre' NaN reikšmes pakeiskite į 'pop'

data['Genre'] = data['Genre'].fillna('pop')
data
```

```python
# 9 Stulpelyje 'Popularity' trūkstamas reikšmes pakeiskite į stulpelio vidurkį

data["Popularity"] = data["Popularity"].fillna(value=data["Popularity"])
```
